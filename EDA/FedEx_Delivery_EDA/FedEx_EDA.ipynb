## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1. Convert Date Columns to datetime format
date_columns = ['PQ First Sent to Client Date', 'PO Sent to Vendor Date',
                'Scheduled Delivery Date', 'Delivered to Client Date',
                'Delivery Recorded Date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# 2. Convert numeric columns stored as object to numeric
numeric_cols = ['Weight (Kilograms)', 'Freight Cost (USD)']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col].str.replace(',', '').replace('See ASN-93 (ID#:1281)', np.nan), errors='coerce')

# 3. Handle missing values
# Fill categorical columns with 'Unknown'
categorical_cols = df.select_dtypes(include='object').columns
df[categorical_cols] = df[categorical_cols].fillna('Unknown')

# Fill numeric columns with 0
numeric_cols_all = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols_all] = df[numeric_cols_all].fillna(0)

# 4. Remove duplicates if any
df = df.drop_duplicates()

# 5. Create Delivery Delay column (days)
df['Delivery Delay (Days)'] = (df['Delivered to Client Date'] - df['Scheduled Delivery Date']).dt.days

# 6. Create On-Time Delivery Flag
df['On_Time_Delivery'] = np.where(df['Delivery Delay (Days)'] <= 0, 1, 0)

# 7. Standardize column names (lowercase, underscores)
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('/', '_').str.lower()

# Check first few rows
df.head()
````
